```python
import pandas as pd
import matplotlib.pyplot as plt

#sample data
data = {
    'Name' : ['Alice', 'Bob', 'Charlie', 'Diana', 'Ethan', 'Fiona', 'George', 'Hannah', 'Ian', 'Julia'],
    'Gender' : ['Female', 'Male', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
    'Age' : [25, 32, 28, 22, 35, 30, 40, 27, 45, 29],
    'Salary' : [55000, 65000, 70000, 48000, 80000, 62000, 90000, 58000, 100000, 67000]
    }
df = pd.DataFrame(data)
```

```python
# Bar Chart for Gender Distribution

plt.figure(figsize = (8, 5))
df['Gender'].value_counts().plot(kind = 'bar', color = 'skyblue')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Number of people')
plt.show()
```

```python
# Histogram for Age Distribution

plt.figure(figsize = (8, 5))
plt.hist(df['Age'], bins = 5, color = 'lightgreen', edgecolor = 'black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Histogram for Salary Distribution

plt.figure(figsize = (8, 5))
plt.hist(df['Salary'], bins = 5, color = 'darkorchid', edgecolor = 'black')
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
# Calculate the average salary by gender
avg_salary_by_gender = df.groupby('Gender')['Salary'].mean()

# Plot the bar chart
plt.figure(figsize=(8, 5))
avg_salary_by_gender.plot(kind='bar', color=['red', 'blue'], edgecolor='black')
plt.title('Average Salary by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Salary')
plt.xticks(rotation=0)
plt.show()
```

```python
# Calculate the average salary by age
avg_salary_by_gender = df.groupby('Age')['Salary'].mean()

# Plot the bar chart
plt.figure(figsize=(8, 5))
avg_salary_by_gender.plot(kind='bar', color='blue', edgecolor='black')
plt.title('Average Salary by Age')
plt.xlabel('Age')
plt.ylabel('Average Salary')
plt.xticks(rotation=0)
plt.show()
```
